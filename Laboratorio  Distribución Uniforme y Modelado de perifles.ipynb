# Distribución Uniforme y Modelado de perifles

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#directorios
path = "C:/Users/Nicole/Desktop/MLpythonModelado"
file = "datoscompletos2020perfiles.csv"
pathdata = os.path.join(path,file)
```

```python
#Lectura CSV
data = pd.read_csv(pathdata)
#Impresion del Csv
data.head(25)
```

# Histograma para mostrar el salario máximo y  el salario mínimo

```python
ingresos = data["ingresos"]
#print(ingresos)
IngresoMayor = max(ingresos, key=int)
IngresoMenor = min(ingresos, key=int)

print('Salario mayor es: ', IngresoMayor)
print('Salario menor es: ', IngresoMenor)
```

```python
X = list(["Salario Mayor", "Salario Menor"])
Y = list([IngresoMayor, IngresoMenor])
fig, axs = plt.subplots(figsize=(10,5),sharey=True)
axs.bar(X,Y)
plt.ylabel("Salario")
plt.title("Salarios")
```

# Estadísticas de los datos

```python
cantdatos=25
```

```python
edad = np.random.uniform(20,40,cantdatos)+ 4*np.random.randn(cantdatos)
```

```python
aniosexp = np.random.uniform(2,20,cantdatos)
```

```python
aniosestudio =np.random.uniform(2,5,cantdatos)
```

```python
datos = pd.DataFrame(
    {
        
        "Edad": edad,
        "Ingresos": ingresos,
        "Año Experiencia": aniosexp,
        "Año Estudio": aniosestudio
    }
)
```

```python
datos.head()
```

```python
#group  by
groupby_edad= datos.groupby("Edad")
```

```python
#Agruptación doble
doble_agrupación = datos.groupby(["Año Experiencia","Año Estudio"])
```

```python
#Estadistíca de todos los datos
doble_agrupación.describe()
```

```python
# En la primera columna se puede observar el año de experiencia donde el más bajo es de 2.035042 y el más alto de 29.540747, 
#hora bien, en la segunda columna están los años de estudio las cuales están en un rangp de 2 - 4 años respectivamente. 
```

# Diagramas de dispersión para verificar sí la edad tiene influencia con los ingreso

```python
sns.set_theme(color_codes=True)
sns.regplot(x="Edad", y="Ingresos", data=datos);
```

```python
# Posiblemente haya una influencia entre la edad y el ingreso pero no es sufiecientemente positiva que tenga 
# una relacion con el ingreso

#Ingreso según los resultados gráficos podemos observar que existe una relación muy poca y casi nula 
#entre la edad y los ingresos, esto puede ser afectado por algunas de las profesiones en donde se prefiere usar 
#personas jóvenes
```

# Diagrama de experiencia tienen que ver con los ingresos

```python
sns.set_theme(color_codes=True)
sns.regplot(x="Año Experiencia", y="Ingresos", data=datos);
```
